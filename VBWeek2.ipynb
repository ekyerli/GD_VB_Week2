```python
#Write a function that returns an array of ones with zeros where both row and column numbers are even. Sample array is below. Number of rows and columns will be entered as parameters.
def matrisolustur(sutun,satır):
    liste = [[0 for col in range(satır)] for row in range(sutun)]
    liste = np.array(liste)
    for i in range(sutun):
        for j in range(satır):
            if i%2!=0 and j%2!=0:
                liste[i][j]= 0
            else:
                liste[i][j]=1
    print(liste)

a=int(input("Satır sayısını giriniz"))
b=int(input("Sutün sayısını giriniz"))
matrisolustur(a,b)

```

# Matplotlib ile Temel Grafikler

```python
#Plot a bar chart for total cases of the 20 countries that havebiggest numbers.
import pandas as pd
import matplotlib.pyplot as plt
covid=pd.read_csv("owid-covid-data.csv", error_bad_lines=False)
grafik=covid.groupby('location').total_deaths.max().sort_values(ascending=False).head(27)
grafik=grafik[7:]
plt.figure(figsize=(16, 5))
plt.plot(grafik)
plt.ylabel('Total Death')
plt.xlabel('Countries')
plt.title('Total Death for Countries')
plt.xticks(rotation = 90, fontsize = 10)
plt.show()
```

```python
#Plot a histogram for daily deaths for any country you choose. Make three subplots for different bins.
grafik=covid[covid['location'].isin(['Turkey'])]
plt.figure(figsize=(16, 5))
plt.hist(grafik.total_deaths)
plt.show()
```

```python
#Plot a scatter plot of new cases and new death for Germany and France.
plt.figure(figsize=(10, 6))
for location in ['France', 'Germany']:
    plt.scatter(covid[covid.location == location].new_cases, 
                covid[covid.location == location].new_deaths, label = location)
plt.xlabel('new_cases', fontdict = {'family': 'Arial', 'color': 'darkblue', 'size': 15 })
plt.ylabel('new_deaths', fontdict = {'family': 'Arial', 'color': 'darkblue', 'size': 15 })
plt.legend()
plt.show()
```

```python
#Plot a boxplot for daily deaths for any country you choose.
grafik=covid[covid['location'].isin(['Turkey'])]
plt.figure(figsize=(10, 5))
plt.boxplot(grafik['new_deaths'].dropna())
plt.show()
```

```python
#Calculate the total case for each continent and plot a pie chart
grafik=covid[covid['location'].isin(['Asai','Africa','Europe','North America','North America','Oceania'])]
continents = [ 'Africa','Asia', 'Europe', 'North America','Oceania', 'South America']
grafik=covid.groupby('continent').sum()
plt.pie(grafik.total_cases, labels=continents,  autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()

```

# İçerik - Seaborn ile Görselleştirme


```python
#Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

covid = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False, error_bad_lines=False)
veri=covid[covid['location'].isin(['Spain', 'France', 'Germany', 'Italy'])]
veri=veri[veri['date']>"2020-4"]
veri=veri[veri['date']<"2020-8"]
plt.figure(figsize=(12, 5), dpi = 500)
plt.title("Total Deaths")

sns.lineplot(x = veri.date, y =veri.total_deaths,hue = 'location', data = veri,err_style = None )
plt.show()
```

```python
#Plot a bar plot with seaborn for average death number that compares continents.
plt.figure(figsize=(12, 5), dpi = 500)
sns.barplot(x='continent',y = 'total_deaths', data = covid)
plt.show()
```

```python
#Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and kde arguments.
grafik=covid[covid['location'].isin(['Turkey'])]

plt.figure(figsize=(12, 4), dpi = 300)

plt.subplot(1,4,1)
sns.distplot(grafik.new_deaths.dropna(), kde = False, bins=20)
plt.title("kde = False bins=20 ")

plt.subplot(1,4,2)
sns.distplot(grafik.new_deaths.dropna(), kde = False, bins=50)
plt.title("kde = False bins=50 ")

plt.subplot(1,4,3)
sns.distplot(grafik.new_deaths.dropna(), bins=20)
plt.title("kde = True bins=20 ")


plt.subplot(1,4,4)
sns.distplot(grafik.new_deaths.dropna(), bins=50)
plt.title("kde = True bins=50")



plt.show()
```

```python
#Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose
grafik=covid[covid['location'].isin(['Turkey','Germany'])]


plt.figure(figsize=(12,6), dpi = 200)

plt.subplot(1,3,1)
sns.barplot(x = 'location', y = 'new_deaths', data = grafik , hue='location')
plt.title("boxplot ")

plt.subplot(1,3,2)
sns.violinplot(x = 'location',  y = 'new_deaths', data = grafik,hue='location', palette = ['#22B573', '#FF914D'])
plt.title("violinplot")
plt.subplot(1,3,3)
sns.swarmplot(x="location", y="new_deaths", data=grafik,hue='location')
plt.title("swarmplot")
```

# İçerik - Plotly Express ile Görselleştirme

```python
#Plot a line plot with plotly express for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px


covid = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False, error_bad_lines=False)


veri=covid[covid['location'].isin(['Spain', 'France', 'Germany', 'Italy'])]
veri=veri[veri['date']>"2020-4"]
veri=veri.groupby(veri.location).sum()
fig = px.line(veri, y='total_deaths', title='Total Deaths')
fig.show()
```

```python
#Plot a bar plot with plotly express for average death number that compares continents.
veri=covid.groupby(covid.continent).mean()

fig = px.bar(data_frame=veri, 
             y='total_deaths',color="total_deaths")
fig.show()
```

```python
#Plot a histogram for daily deaths for any country you choose.
grafik=covid[covid['location'].isin(['Turkey'])]
fig = px.histogram(grafik, x="new_deaths")
fig.show()
```
